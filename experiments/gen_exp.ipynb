```python
dict_d = dict(zip(DCFS, DCMS))
dict_o = dict(zip(OBJECT_RANKERS, OR_MODELS))
del dict_d[PCL]
models_dict = {OBJECT_RANKING: dict_o, DISCRETE_CHOICE: dict_d}
metric_name_dict = {OBJECT_RANKING: '0/1 Ranking Accuracy', DISCRETE_CHOICE: 'NormalizedAccuracy', CHOICE_FUNCTION: 'F1Score'}
objects = "Objects"
MODEL = "aModel"
DIR_PATH = os.getcwd()
RESULT_FOLDER = 'results'
datasets = ['medoid','hypervolume']
LPS = [OBJECT_RANKING, DISCRETE_CHOICE]
```

```python
for learning_problem in LPS:
    for dc in datasets:
        df_path = os.path.join(DIR_PATH, RESULT_FOLDER, "generalization_{}_{}.csv".format(dc, learning_problem))
        dataFrame = pd.read_csv(df_path, index_col=0)
        dataFrame.sort_index(axis=1, inplace=True)
        if learning_problem == OBJECT_RANKING:
            vals = dataFrame['err']
            del dataFrame['err']
            dataFrame['err'] = vals
        dataFrame.to_csv(df_path)
dataFrame = pd.read_csv(df_path, index_col=0)
dataFrame.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
sns.set(color_codes=True)
plt.style.use('default')
def get_max_min(maxi, mini, acc):
    if np.max(acc) > maxi:
        maxi = np.max(acc)
    if np.min(acc) < mini:
        mini = np.min(acc)
    return maxi, mini
def get_accuracy(acc):
    if learning_problem==DISCRETE_CHOICE:
        mins = 1/N_OBJECTS_ARRAY
        acc = (acc - mins)/(1.0 - mins) 
    elif learning_problem==OBJECT_RANKING:
        acc = 1 - acc
    return acc
n_objs = 20
```

```python
markers = ['.', '^', 'v', 'x',  'o', "*", "d","P", "8", "s", "H","+"]
#colors = ["darkviolet","darkorange", "green", "red", "grey", "azure",'pink', "darkgrey","aqua","orchid","black","black"]
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
colors = colors + ["black"]
len(colors), len(DCFS)
extension = "pdf"
```

```python
for learning_problem in LPS:
    y_label= metric_name_dict[learning_problem]
    x_label="Number of Objects"
    plot_file = os.path.join(DIR_PATH, RESULT_FOLDER, "generalization_{}.{}".format(learning_problem, extension))
    maxi = 0
    mini = 100
    fig, axs = plt.subplots(1, 2, figsize=(15,7) ,frameon=True, edgecolor='k', facecolor='white')
    for sub_plot, dc in zip(axs, datasets):
        df_path = os.path.join(DIR_PATH, RESULT_FOLDER, "generalization_{}_{}.csv".format(dc, learning_problem))
        dataFrame = pd.read_csv(df_path, index_col=0)
        if learning_problem==DISCRETE_CHOICE:
            dataFrame['feta_dc'] = dataFrame['feta_dc_zero'] 
            del dataFrame['feta_dc_zero']
        if learning_problem==OBJECT_RANKING: 
            dataFrame['feta_ranker'] = dataFrame['feta_ranker_zero'] 
            del dataFrame['feta_ranker_zero']
        models = models_dict[learning_problem]
        print(models)
        for i, (func, model) in enumerate(models.items()):
            alpha = 1.0
            color = colors[i]
            markersize = 6
            if "random" in func:
                color = "black"
                alpha = 0.1
            if func in ["err", "listnet"]:
                continue
            if "ranknet" == func:
                model=model+"/ListNet"
            if "ranksvm" == func:
                model=model+"/ERR"
            acc = get_accuracy(dataFrame[func].values)
            sub_plot.plot(N_OBJECTS_ARRAY[0:n_objs], acc[0:n_objs], label=model, marker=markers[i], 
                          markersize=markersize, color=color,alpha=alpha)
            maxi, mini = get_max_min(maxi, mini, acc)
        sub_plot.set_ylabel(y_label)
        sub_plot.set_xlabel(x_label)
        sub_plot.set_title(dc.title(), horizontalalignment='center', verticalalignment='bottom')
        sub_plot.set_xticks(np.arange(3, N_OBJECTS_ARRAY[0:n_objs][-1], 2))
        sub_plot.set_yticks(np.arange(mini-0.1,maxi+0.1,0.05))
    sub_plot.legend(loc='upper center', ncol=3, fancybox=True, shadow=True)
    plt.show()
    fig_param = {'facecolor':'w', 'edgecolor':'w', 'transparent':False, 'dpi':800, 'format':extension,'bbox_inches':'tight', 'pad_inches':0.05}
    fig_param['fname']=plot_file
    plt.savefig(**fig_param)
```

```python
from sklearn.utils import check_random_state
import os
import numpy as np
from scipy.spatial.distance import squareform, pdist

plot_file = os.path.join(os.getcwd(), "results", "medoid.svg")

seed = 50
n_features = 2
n_objects = 5
n_instances = 2
random_state = check_random_state(seed=seed)
X = random_state.uniform(size=(n_objects, n_features))
Y = np.empty(n_objects)
D = squareform(pdist(X, metric='euclidean'))
sum_dist = D.mean(axis=0)
medoid = np.argmin(sum_dist)
Y = np.argsort(np.argsort(D[medoid]))
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
from matplotlib.font_manager import FontProperties
font = FontProperties()
font.set_size('large')
font.set_family('sans-serif')
font.set_weight('medium')
sns.set(color_codes=True)
plt.style.use('default')
#sns.set_style("dark")
fig, ax = plt.subplots(figsize=(5, 5), dpi=80, facecolor='w', edgecolor='w')
ax.scatter(X[:, 0], X[:, 1])
labels = ['A', 'B','C','D','E']
for i in range(n_objects):
    ax.text(X[i, 0]+0.02,
            X[i, 1]-0.01,
            s=labels[Y[i]], color='k',fontproperties=font)
plt.plot(X[medoid, 0], X[medoid, 1], 'ro')
#for x in X:
    #plt.plot([X[medoid, 0],x[0]], [X[medoid, 1],x[1]], color="k",linewidth=0.8)
plt.xticks([])
plt.yticks([])
plt.show()
plt.savefig(plot_file, dpi=None, facecolor='None', edgecolor='None', orientation='portrait', format='svg')
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib.font_manager import FontProperties
from sklearn.datasets.samples_generator import make_blobs
%matplotlib notebook
sns.set(color_codes=True)
plt.style.use('default')
plt.rc('text', usetex=True)
plt.rc('font', family='cursive', size=20, weight='heavy')
#sns.set_style("dark")
plot_file = os.path.join(os.getcwd(), "results", "clustering.svg")
fig = plt.figure(figsize=(8, 5), dpi=80, facecolor='None', edgecolor='None')
N = 5000
rs = np.random.RandomState(42)
k=4
X, y = make_blobs(n_samples=N, centers=k, n_features=2, random_state=0, center_box=(-10.0, 10.0), cluster_std=0.9)
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
colors = ['r', 'g', 'b', 'c']
for i in range(4):
    idx = np.where(labels==i)[0]
    txt = r'$c_{}$'.format(i+1)
    plt.scatter(X[idx, 0], X[idx, 1], marker='o', color=colors[i], s=2)
    plt.text(centers[i, 0]+0.12, centers[i, 1]+0.12, s=str(txt), color='k',
             bbox=dict(facecolor='snow', edgecolor='None', pad=0))
plt.scatter(centers[:, 0], centers[:, 1], marker='o', color='k', s=15)
plt.xticks([])
plt.yticks([])
plt.show()
plt.savefig(plot_file, dpi=None, facecolor='None', edgecolor='None', orientation='portrait', format='svg')
```
